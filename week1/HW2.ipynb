```python
import numpy as np

D = np.array( [[0.1,0.2,0.3],[0.3,0.4,0.5],[0.1,0.3,0.2],[0.6,0.7,0.8],[0.8,0.5,0.4],[0.9,0.1,0.2]] )

Dh = np.hstack((D, np.ones([6,1])))

print(D)
print(Dh)
```

```python

t = np.identity(4)
t[:,3] = [-0.5,0.5,0.1,1]
print(t)

t@Dh.T
```

```python
S = np.identity(4)
row, col = np.diag_indices_from(S)
S[row,col] = np.array([3, 2, 4, 1])
print(S)
```

```python
min_i = D.min(axis = 0)
mx_i = D.max(axis = 0)
T = np.identity(4)
T[:3,3] = -min_i
print(T)

scale = 1/(mx_i-min_i)
print(scale)
S = np.identity(4)
row, col = np.diag_indices_from(S)
S[row,col] = np.append(scale,1)
print(S)
result = S@T@Dh.T
print(result.T)
```

```python
mins = D.min(axis = 0)
T[:3,3] = -mins
print(T)
S = np.eye(4)
maxes = D.max(axis = 0)
print("maxes",maxes)
scale = 1/(maxes-mins)
print("scale", scale)
row, col = np.diag_indices_from(S)

S[row,col] = np.append(scale,1)
print("scaling matrix:", S)

result = S@T@Dh.T
result = result[:D.shape[1],:]
result = result.T
print (result)
```
